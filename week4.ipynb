```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the 'tips' dataset
tips = sns.load_dataset("tips")
print(tips.head())
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs. Tip by Gender")
plt.show()
sns.histplot(tips["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.show()
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
sns.boxplot(data=tips, x="day", y="tip_pct", hue="sex")
plt.title("Tip Percentage by Day and Gender")
plt.show()
sns.countplot(data=tips, x="day", hue="smoker")
plt.title("Customer Count by Day and Smoking Status")
plt.show()
```
